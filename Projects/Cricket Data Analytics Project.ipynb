```python
with open ('/content/t20_wc_batting_summary.json') as f:
   batting_data = json.load(f)

   all_records = [] ##single list


   for record in batting_data:
    all_records.extend(record['battingSummary'])

df_batting = pd.DataFrame(all_records)  #crate data frame
df_batting.head(11)
```

```python
#create new columns

df_batting["out/not_out"] = df_batting.dismissal.apply(lambda x :'out' if len(x)>0 else "not_out")
df_batting.head(11)
```

```python
#drop the columns

df_batting.drop(columns=['dismissal'], inplace = True)
df_batting.head(11)
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x : x.replace('a$',''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x : x.replace('\xa@',''))
df_batting.head(11)
```
